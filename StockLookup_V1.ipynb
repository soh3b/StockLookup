```python
import yfinance as yf
import pandas as pd
import seaborn as sns
from datetime import date
import dateutil.relativedelta
import matplotlib.pyplot as plt



#setting up today and last month
today = date.today()
lastMonth = today + dateutil.relativedelta.relativedelta(months=-1)
print(str(today))
print(str(lastMonth))

      

    
```

```python
#Getting the ticker symbol and verifying it 
tick = input("Enter a symbol: ")
verified = False
while verified == False:
    for v in tick.split():
        if v.isdigit():
            print("This is not a valid symbol please try again")
            tick = input("Enter a symbol: ")
        else: verified = True
```

```python
td = yf.Ticker(tick.upper())

tickerDf = td.history(period='1mo')
close=tickerDf['Close']
prices_list = [round(val, 2) for val in close.tolist()]

```

```python
sns.lineplot(data= prices_list )
sns.set_theme()
plt.xticks(rotation=30)
plt.title(f"Closing Stock Prices for {tick}")
plt.show()
```
