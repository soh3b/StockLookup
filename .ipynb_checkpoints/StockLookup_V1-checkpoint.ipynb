```python
#Getting the ticker symbol and verifying it 
tick = input("Enter a symbol: ")
verified = False
while verified == False:
    for v in tick.split():
        if v.isdigit():
            print("This is not a valid symbol please try again")
            tick = input("Enter a symbol: ")
        else: verified = True
```

```python
td = yf.Ticker(tick.upper())

tickDf = td.history(period='1mo')
close=tickDf['Close']


```

```python
tickDf.plot(kind="line",figsize=(12,12),subplots = True)
```
